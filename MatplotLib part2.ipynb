```python
# pie chart-----> 
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = [34,45,67,89]
mylabel = ['A','B','C','D']
plt.pie(a,labels = mylabel)
```

```python
a = [34,45,67,89]
mylabel = ['A','B','C','D']
plt.pie(a,labels = mylabel,autopct = '%.3f')
```

```python
# subplot-----> if u want to create more than one graph and try
# to compare.

x = np.array([23,45,78,90])
y = np.array([23,89,23,7])
plt.subplot(2,1,1)
plt.plot(x,y,color = 'r')

a = np.array([23,45,78,90])
b = np.array([67,89,23,45])
plt.subplot(2,1,2)
plt.plot(a,b,'+' )
```
