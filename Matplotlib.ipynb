```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line plot
# Histogram
# barchart
# pie chart
# subplot

# line plot
no_of_classes = np.array([2,4,6,8])
no_of_student = np.array([4,6,8,9])
plt.plot(no_of_classes,no_of_student)
```

```python
no_of_classes = np.array([2,4,6,8])
no_of_student = np.array([4,6,8,9])
plt.xlabel('Number of classes')
plt.ylabel('Number of students')
plt.plot(no_of_classes,no_of_student)
```

```python
no_of_classes = np.array([2,4,6,8])
no_of_student = np.array([4,3,2,1])
plt.xlabel('Number of classes')
plt.ylabel('Number of students')
plt.plot(no_of_classes,no_of_student)
```

```python
# temp
num_of_days = [2,4,6,8,12,20]
temp_in_cel = [45,23,20,12,34,42]
plt.xlabel('Number of days')
plt.ylabel('Temperature')
plt.plot(num_of_days,temp_in_cel)
```

```python
# bar plot
subject = ['AI','ML','DBMS',"Python"]
day1 = [5,10,12,4]
day2 = [6,7,2,12]
day3 = [10,18,30,20]
plt.xlabel('Subjects')
plt.ylabel('number of students')
plt.bar(subject,day1,color='r')
```

```python
# horizontal bar plot
subject = ['AI','ML','DBMS',"Python"]
day1 = [5,10,12,4]
day2 = [6,7,2,12]
day3 = [10,18,30,20]
plt.xlabel('Subjects')
plt.ylabel('number of students')
plt.barh(subject,day1,color='g')
```
